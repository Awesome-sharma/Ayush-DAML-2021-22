## EXPLORATORY DATA ANALYSIS

### State with highest to lowest no. of accidents.

```python
Total_Persons_accidents = pd.DataFrame(RA.groupby('States/UTs').sum()['Total - Total Accidents'])
Total_Persons_accidents.sort_values(['Total - Total Accidents'], ascending = False)
```

```python
# State with highest to lowest no. of accidents. --- Visulization ---
RA_TA = RA['Total - Total Accidents']
ax = RA_TA.plot.barh(figsize=(20, 20) )
```

### Total no. of accidents in 2018 all over India.

```python
Total_accidents=RA.sum(axis = 0, skipna = True)
print("\nTotal no. of accidents in 2018 all over India -> ",Total_accidents['Total - Total Accidents'],"\n")
```

### Total no. of persons killed in accidents in 2018 all over India.

```python
Total_accidents=RA.sum(axis = 0, skipna = True)
print("\nTotal no. of persons killed in accidents in 2018 all over India -> ",Total_accidents['Total - Persons Killed'],"\n")
```

### States/UTs with lowest to highest no. of killed people in accidents.

```python
Total_Persons_Killed_in_StatesUTs = pd.DataFrame(RA.groupby('States/UTs').sum()['Total - Persons Killed'])
Total_Persons_Killed_in_StatesUTs.sort_values(['Total - Persons Killed'])
```

```python
#States/UTs with lowest to highest no. of killed people in accidents. --- Visulization ---
RA_TK = RA['Total - Persons Killed']
ax = RA_TK.plot.barh(figsize=(20, 20) )
```

### States/UTs wise Total Persons Killed in Total Accidents 

```python
cat_subcat = pd.DataFrame(RA.groupby(['States/UTs', 'Total - Total Accidents']).sum()['Total - Persons Killed'])

# Sorting the values
cat_subcat.sort_values(['Total - Total Accidents'])
```

### Difference between Total Accidents and Total Persons Killed

```python
RA1 = RA
RA1['Total Accidents - Persons Killed'] = (RA['Total - Total Accidents'] - RA['Total - Persons Killed'])

cat_subcat = pd.DataFrame(RA1.groupby(['States/UTs', 'Total - Total Accidents','Total - Persons Killed']).sum()['Total Accidents - Persons Killed'])

# Sorting the values
cat_subcat.sort_values(['Total Accidents - Persons Killed'])
```
